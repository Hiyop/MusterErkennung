```python
import numpy as np
import matplotlib.pyplot as plt

def getData():
    file = open("res/zip.train", "r") 
    VecData=[]
    LabelData = []
    for line in file:
        tmpLine = line.split()
        floatLine = []
        Linelen = len(tmpLine)            
        for i in range(1,Linelen):
            floatLine.append(float(tmpLine[i]))
        VecData.append(floatLine)
        LabelData.append(float(tmpLine[0]))
    file.closed   
    return VecData, LabelData

def MittelPunktVector(VecList):    
    M = np.array(VecList[0])
    n = len(VecList)    
    for i in range(1,n):
        tmpVec = np.array(VecList[i])
        M = M + tmpVec
    M = M / (n) 
    return M

def getKovarianzMatrix(Data, M):  
    n = len(Data)    
    x = np.matrix(Data[0])
    X = np.transpose((x - M)) @ (x - M)
    for i in range(1,n):        
        x = np.matrix(Data[i])
        tmpX = np.transpose((x - M)) @ (x - M)
        X = X + tmpX 
    return (X / n)

vecs, label = np.array(getData())
m = MittelPunktVector(vecs)
kovmat = getKovarianzMatrix(vecs, m)

eigenvalues, eigenvectors = np.linalg.eig(kovmat)

inds = eigenvalues.argsort()

#print(inds[-2:])



eigenvalues = eigenvalues[inds]

eigenvectors = eigenvectors[inds][-2:]
#print()
print(eigenvectors[1][0:2])

newVecs = []

for i in range(0, len(vecs)):
    tmpVec = np.array([vecs[i][inds[-2:][0]] ,vecs[i][inds[-2:][1]]])   
    newVecs.append(tmpVec @ np.transpose(eigenvectors))

print(newVecs[0])
  
newVecs = np.array(newVecs)   
plt.scatter(x,y)
plt.show()
```
