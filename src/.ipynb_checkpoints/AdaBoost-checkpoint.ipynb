```python
import os

posWords = ["Amazing", "Awesome", "Excellent", "Fabulous", "Fantastic", "Great", "Incredible", "Outstanding", "Perfect", "Remarkable", "Smart", "Spectacular", "Splendid", "Stellar", "Super", "Ultimate", "Unbelievable", "Wondrous"]

def getList(path):
    words = []
    for filename in os.listdir(path):
        #path = os.path.join(path, filename)
        file = open(path + filename, "r", encoding="utf8")
        string = file.read()
        string = string.replace("<br /><br />", " ")
        string = string.replace("'", "")
        string = string.replace(".", "")
        string = string.replace(",", "")
        string = string.replace(":", "")
        string = string.replace('"', "")
        string = string.replace("!", "")
        string = string.replace("(", "")
        string = string.replace(")", "") 
        string = string.upper()
        words += string.split()  

    return words
    #counts = dict()

    #for word in words:
    #    if word in counts:
    #        counts[word] += 1
    #    else:
    #       counts[word] = 1

    #return sorted(counts.items(), key=lambda item: item[1], reverse=True)
    
def count(sequence, item):
    cont=0
    for i in sequence:
        if item==i:
            cont+=1
    return cont

posList = getList("res/aclImdb/train/pos/")
negList = getList("res/aclImdb/train/neg/")

for word in posWords:
    print(word)
    print(str(count(posList, word.upper())))
    print(str(count(negList, word.upper())))
    
    
```
