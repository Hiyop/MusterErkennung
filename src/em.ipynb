## Hilfsfunktionen

- mahalanobisDistanz(x, mu, sigma): Berechnung der Mahalanobisdistanz zwischen Vektor $\vec x$ und Mittelpunkt $\vec mu$ mit Kovarianzmatrix $\Sigma$

$$d = (\vec x - \vec \mu)^T \Sigma^{-1} (\vec x - \vec \mu)$$

- MittelPunktVector(VecList): Berechnung des Mittelpunktvektors aus einer Liste von Vektoren $\vec v_0$ bis $\vec v_N$  
  
$$\vec M = \frac{1}{N} \sum_{i = 0}^N \vec v_i $$  
  
- getKovarianzMatrix(Data, M): Berechnung der Kovarianzmatrix $\Sigma$ aus einer Liste von Vektoren $\vec v_0$ bis $\vec v_N$ und zugehörigem Mittelpunktvektor $\vec M$  
  
$$\Sigma = \frac{1}{N}\sum_{i = 0}^N (\vec v_i - \vec M)(\vec v_i - \vec M)^T $$  

```python
def mahalanobisDistanz(x, mu, sigma):
    return np.matrix(x-mu) @ np.linalg.pinv(sigma) @ np.transpose(np.matrix(x-mu))

def MittelPunktVector(VecList):    
    n = len(VecList)    
    M =  np.sum(VecList , axis=0)
    M = M / (n) 
    return M

def getKovarianzMatrix(Data, M):  
    n = len(Data)    
    x = np.matrix(Data[0])
    X = np.transpose((x - M)) @ (x - M)   
    for i in range(1,n):        
        x = np.matrix(Data[i])
        tmpX = np.transpose((x - M)) @ (x - M)        
        X = X + tmpX        

    return (X / n)
```

## EM Clustering

- initialisiere sigma mit Identitätsmatrix und mu mit zufalls Punkten aus dem Datensatz
- iteriere über jeden Datenpunkt und berechne für jeden Cluster den Mahalanobisabstand zum zugehörigen Mittelpunkt
- weise den jeweiligen Datenpunkt dem Cluster mit der kleinsten Abstand zu
- iteriere über jedes Cluster und berechne die neue jeweilige Kovarianzmatrix und den Mittelpunkt
- wiederhole das ganze für die gegebene Anzahl von attempts
- berechne die Güte des Clusterings mit Hilfe der Determinanten
$$\sum_{i = 0}^{\#cluster} |\Sigma_i| $$

```python
def em_clustering(data, clusters, attempts):
    
    sigmaList = []
    muList = []    
    clusterList = []   
    numData = len(data)
    
    for i in range(0,clusters):
        sigma = np.identity(2)
        mu = np.array(data[np.random.randint(0, len(data))])        
        sigmaList.append(sigma)
        muList.append(mu)

    for i in range(0,attempts):
        
        clusterList = [[] for x in range(clusters)]
        
        
        for j in range(0,len(data)):

            distList = []
            
            for k in range(0, clusters):                
                distList.append(mahalanobisDistanz(data[j], muList[k], sigmaList[k]))

            targetCluster = 0
            minDist = distList[0]

            for l in range(1, len(distList)):
                if(minDist > distList[l]):
                    minDist = distList[l]
                    targetCluster = l
            
            clusterList[targetCluster].append(data[j])            
            
        for m in range(0,clusters):
            if(len(clusterList[m]) > 0):
                muList[m] = MittelPunktVector(clusterList[m])    
                sigmaList[m] = getKovarianzMatrix(clusterList[m], muList[m])         
       

    quality = 0
    for sigma in sigmaList:
        quality += np.linalg.det(sigma)
    
    #print(quality)
    
    return (quality, clusterList, muList, sigmaList)
```

## Ausführung des Clusterings

- führe Clustering 10 mal aus und behalte jeweils das Ergebniss mit der höchsten Güte

```python
def exec_clustering( cluster ):
    
    if(cluster > len(colorList)):
        print("Nicht genügend Farben für Cluster")    

    vecList = np.genfromtxt('res/2d-em.csv', delimiter=',')
    
    quality, clusterList, muList, sigmaList = em_clustering(vecList, cluster, 10)

    for i in range(0,10):
        tmp_quality, tmp_clusterList, tmp_muList, tmp_sigmaList = em_clustering(vecList, cluster, 10)
        if(quality > tmp_quality):        
            quality, clusterList, muList, sigmaList = tmp_quality, tmp_clusterList, tmp_muList, tmp_sigmaList
    
    
    #Ploting
    for i in range(0,cluster):
        if(len(clusterList[i]) > 0 ):
            x, y = np.array(clusterList[i]).T            
            plt.plot(x,y, colorList[i])

    num_points = 1000
    radius = 1
    arcs = np.linspace(0, 2 * pi, num_points)
    x_circle = radius * sin(arcs)
    y_circle = radius * cos(arcs)


    for i in range(0, cluster):   
        xy = np.array(list(zip(x_circle, y_circle)))
        conv = sigmaList[i]
        mu = muList[i]

        xy = (xy @ conv) + mu

        x, y = xy.T

        plt.plot(x,y, 'k,')

    plt.show()
    print("Güte: " + str(quality))
    return
```

```python
exec_clustering(2)

exec_clustering(3)

exec_clustering(4)

exec_clustering(5)

exec_clustering(6)
```
