```python
import os

posWords = ["Amazing", "Awesome", "Excellent", "Fabulous", "Fantastic", "Great", "Incredible", "Outstanding", "Perfect", "Remarkable", "Smart", "Spectacular", "Splendid", "Stellar", "Super", "Ultimate", "Unbelievable", "Wondrous"]
negWords = ["boring", "dull", "expected","bad","failing","imperfect", "indistinguished","inferior","poor", "second-class", "second-rate", "unnoteworthy","unworthy"]

def getList(path):
    words = []
    for filename in os.listdir(path):
        #path = os.path.join(path, filename)
        file = open(path + filename, "r", encoding="utf8")
        string = file.read()
        string = string.replace("<br /><br />", " ")
        string = string.replace("'", "")
        string = string.replace(".", "")
        string = string.replace(",", "")
        string = string.replace(":", "")
        string = string.replace('"', "")
        string = string.replace("!", "")
        string = string.replace("(", "")
        string = string.replace(")", "") 
        string = string.upper()
        words += string.split()  
        file.close()        
    return words

def countData(path):
    return len(os.listdir(path))
    #counts = dict()

    #for word in words:
    #    if word in counts:
    #        counts[word] += 1
    #    else:
    #       counts[word] = 1

    #return sorted(counts.items(), key=lambda item: item[1], reverse=True)
    
def count(sequence, item):
    cont=0
    for i in sequence:
        if item==i:
            cont+=1
    return cont

num = countData("res/aclImdb/train/pos/")
num += countData("res/aclImdb/train/neg/")

posList = getList("res/aclImdb/train/pos/")
negList = getList("res/aclImdb/train/neg/")

print("Num " + str(num))
 
for word in posWords:    
    countPos = count(posList, word.upper())       
    print(word)
    print(countPos)
    print(str(countPos/num))
        
    
    
#for word in negWords:
#    countPos = count(posList, word.upper())
#    countNeg = count(negList, word.upper())
#    print(word)
#    print(str(countPos))
#    print(str(countNeg))    
```
